```python
arr3 = np.random.randn(5)
arr4 = np.random.randn(5)
np.add(arr3, arr4)                   # 2ta massivning mos elementlarini qushadi
arr3 + arr4                 #bu ham shu ishni qiladi
```

```python
#Massivlar
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])   #xarr
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])   #yarr
#shart
cond = np.array([True, False, True, True, False])
#where funksiyasi
result = np.where(cond, xarr, yarr)        # cond==True : xarr elementi, cond==False : yarr elementi
result
```

```python
arrwh = np.random.randn(4, 4)
results = np.where(arrwh<0, -2, 2)       # arr ning manfiy elementlarini -2 ga musbat elementlarin esa 2 ga almashitirish
results
```

```python
arrar1 = np.arange(6).reshape(2, 3)            # 0~6 gacha 2 3 massiv
arrar2 = np.arange(6, 12).reshape(2, 3)        # 6~12 gacha 2 3 massiv
arrar2
```

```python
arrar1 + arrar2            # 1-va 2-massivlarni qo'shish
arrar1 * arrar2            # 1-va 2- massivlarni ko'paytirish
arrar1 / arrar2            # 1-va 2- massivlarni  bo'lish
arrar1 - arrar2            # 1-va 2- massivlarni  ayirish
arrar1 ** 2                # 1-massiv elementlarini kvadratga oshiradi   (square)
arrar2 ** 0.5              # 1-massiv elementlarini kvadrat ildizi    (sqrt  )
```

```python
arr_sort = np.random.randn(5)    # 4 ta taxminiy elementlardan tashkil topgan massiv    
# 1-usul
sorted_arr = np.sort(arr_sort)      # berilgan massivdan nusxa olib tartiblash yani b-n massiv o'zgarmaydi

# 2-usul
arr_sort.sort()          # b-n massivning uzida tartiblash yani b-n massiv o'zgaradi

# teskari tartiblash
rev_arr = -np.sort(-arr_sort)             # b-n massivni teskari tartiblash

# 2d massivlarni tartiblash
arr2d = np.random.randn(2, 4)         # (2:4) massiv
# axis = 0 --> Qator
row_str = np.sort(arr2d, axis=0)    
row_str   
# axis = 1 --> Ustun
col_str = np.sort(arr2d, axis=1)
col_str
```

```python
# Takrorlanmas amallar
names = np.array(['Sarvar', 'Abdurahmon', 'Hasan',  'Temur', 'Sarvar', 'Temur'])
np.unique(names)              # names dagi takrorlanmas elementlarni qaytaradi (takrorlansa ham faqat bittasini qabul qiladi)

# in1d usuli
arr1d = np.array([6, 0, 0, 3, 2, 5, 6])
arr2d = np.array([0, 2, 3])
np.in1d(arr1d, arr2d)          # arr2 dagi elementlarni arr1 da mavjudligini tekshirish

# setdiff1d usuli
np.setdiff1d(arr1d, arr2d)         # names1 massivining names2 massividagi takrorlanmas qismini qaytaradi
```
